```python
import pandas as pd
import numpy as np

dapoer = pd.read_csv('D:\Self\Learning Stuff\IndoDapoer\INDODAPOERData.csv')
dapoer.head()
```

```python
dapoer.columns
```

```python
dapoer = pd.melt(dapoer, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name = 'year', value_name = 'amount')
dapoer.head()
```

```python
dapoer.drop(['Country Code', 'Indicator Code'], axis=1, inplace = True)
dapoer.head()
```

```python
dapoer_pivot = dapoer.pivot_table(index=['Country Name', 'year'], columns='Indicator Name', values = 'amount', aggfunc = np.sum).reset_index()
dapoer_pivot.head()
```

```python
dapoer['Indicator Name'].unique()
```

```python
dapoer['Country Name'].unique()
```

```python
#separate country/city from its identifier
dapoer_pivot['Country Type'] = dapoer_pivot['Country Name'].str.split(", ", expand=True)[1]
dapoer_pivot['Country/City'] = dapoer_pivot['Country Name'].str.split(", ", expand=True)[0]


dapoer_pivot.head()
```

```python
dapoer_pivot['Country Type'].unique()
```

```python
dapoer_pivot['Country Type'] = dapoer_pivot['Country Type'].str.replace('Kab.', 'Kab')
dapoer_pivot['Country Type'].unique()
```

```python
dapoer_pivot = dapoer_pivot.drop(['Country Name'], axis = 1)
dapoer_pivot.head()
```

```python
dapoer_filter = dapoer_pivot[['year', 'Country Type', 'Country/City', 
                                    'Average National Exam Score: Junior Secondary Level (out of 100, available only in district level for 2009)',
       'Average National Exam Score: Primary Level (out of 100, available only in district level for 2009)',
       'Average National Exam Score: Senior Secondary Level (out of 100, available only in district level for 2009)',
       'Education function expenditure (in IDR)',
       'Human Development Index',
       'Monthly Per Capita Household Education Expenditure (in IDR)',
       'Net Enrollment Ratio: Junior Secondary (in %)',
       'Net Enrollment Ratio: Primary (in %)',
       'Net Enrollment Ratio: Senior Secondary (in %)',
       'Number of schools at junior secondary level',
       'Number of schools at primary level',
       'Number of schools at Senior Secondary level',
       'Number of Student: Junior Secondary Level (in number of people, 2009 data only)',
       'Number of Student: Primary Level (in number of people, 2009 data only)',
       'Number of Student: Senior Secondary Level (in number of people, 2009 data only)',
       'Number of Teacher: Junior Secondary Level (in number of people, 2009 data only)',
       'Number of Teacher: Primary Level (in number of people, 2009 data only)',
       'Number of Teacher: Senior Secondary Level (in number of people, 2009 data only)',
       'Total Specific Allocation Grant for Education (in IDR Billion)'
                             ]]
#dapoer_filter = dapoer_filter.loc[dapoer_filter['Total Expenditure (in IDR)'] > 0]
dapoer_filter.head()
```
